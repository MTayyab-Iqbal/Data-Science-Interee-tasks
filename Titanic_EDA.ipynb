## **Step# 01**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Step# 02**
## Load the Dataset

```python
# Use the seaborn library to load dataset
df = sns.load_dataset('titanic')
```

## **Step# 04**
## Data Cleaning

### 1. Handle missing values

```python
# Fill the missing vlues of 'age' with their mean
df["age"].fillna(df['age'].mean(), inplace= True)
```

```python
# Fill the missing values of 'embarked' with their mode
df["embarked"].fillna(df['embarked'].mode()[0], inplace= True)

# Fill the missing values of 'embark_town' with their mode
df["embark_town"].fillna(df['embark_town'].mode()[0], inplace= True)
```

```python
# Verify missing values
df.isnull().sum()
```

### 2. Removing column

```python
# Remove the 'deck' column as it has many null values
df.drop('deck', axis=1, inplace=True)
```

```python
# Verify droping column
df.head()
```

### 3. Removing Duplicates[link text](https://)

```python
# Remove the duplicated rows from dataset (if any)
df.drop_duplicates(inplace = True)
```

```python
# Verify the no of rows after removing duplicates
df.shape
```

## **Step# 05**
## Data Visualization

### 1. Bar chart for one categorical variable

```python
# making countplot for survival ones
sns.countplot(x='survived', data=df, palette= 'pastel')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.title('Survival Count')
plt.show()
```

### 2. Histogram for a numerical variable

```python
# making histogram of age
sns.histplot(df['age'], bins= 20, kde= True, color= 'violet')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

### 3. Correlation Heatmap

```python
# Making heatmap of correlation between numerical variables
sns.heatmap(df.corr(), annot= True, cmap= 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
